# Histogram: Create a histogram to visualize the distribution of a single variable. Create histograms for each feature, and color the bars by species. Customize the plot by adding a title, labels for the x and y-axis, and setting the number of bins.

```python
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_data = load_iris()

# Create a DataFrame with the feature data
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the target variable 'species' to the DataFrame
iris_df['species'] = iris_data.target

# Replace the numerical target variable with the actual species names
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Create histograms for each feature, colored by species
for feature in iris_data.feature_names:
    sns.histplot(data=iris_df, x=feature, hue='species', bins=10, kde=True, element='step', stat='density', common_norm=False, palette='Set1')

# Add title and labels
plt.title('Histogram of Iris Dataset Features by Species')
plt.xlabel('Feature Value')
plt.ylabel('Density')

# Specify legend labels
legend_labels = {'setosa': 'Setosa', 'versicolor': 'Versicolor', 'virginica': 'Virginica'}
plt.legend(title='Species', labels=legend_labels)

# Show the plot
plt.show()
```

# Bar Chart: Create a bar chart to visualize the frequency or count of a categorical variable. Create a bar chart for the target variable (species), showing the count of each species. Customize the plot by adding a title, labels for the x and y-axis, and changing the color scheme.

```python
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the target variable 'species' to the DataFrame
iris_df['species'] = iris_data.target

# Replace the numerical target variable with the actual species names
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Count the frequency of each species
species_counts = iris_df['species'].value_counts()

# Create bar chart
sns.barplot(x=species_counts.index, y=species_counts.values, palette='pastel')

# Add title and labels
plt.title('Frequency of Species in Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Count')

# Show the plot
plt.show()
```

# Heatmap: Create a heatmap to visualize the correlation between multiple variables. Create a heatmap showing the correlation matrix between the features. Customize the plot by adding a title and changing the color scheme.

# Dropping Null Values: Remove any rows that have missing values.

```python
# As their are no missing values, I am artificially adding missing values using pandas and mean so that the actual reseblance ramains the same

import pandas as pd
from sklearn.datasets import fetch_california_housing
import numpy as np

# Fetch the California housing dataset
california_housing = fetch_california_housing()

# Convert the dataset into a pandas DataFrame
california_df = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)

# Introduce artificial missing values
np.random.seed(0)  # For reproducibility
missing_mask = np.random.rand(*california_df.shape) < 0.1  # Introduce missing values with a probability of 10%
california_df[missing_mask] = np.nan

# Check if missing values were introduced
print("Missing values introduced:", california_df.isnull().sum())

# Imputation using mean
california_df_imputed = california_df.fillna(california_df.mean())

# Check if missing values were imputed
print("Missing values after imputation:", california_df_imputed.isnull().sum())
```

```python
# Drop rows with missing values
california_df_dropped = california_df.dropna()

# Check if any missing values remain
print("Missing values after dropping rows:", california_df_dropped.isnull().sum())
```

```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Fetch the California housing dataset
california_housing = fetch_california_housing()

# Convert the dataset into a pandas DataFrame
california_df = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)

# Introduce artificial missing values in the "total_bedrooms" column
california_df.loc[2:5, "total_bedrooms"] = None

# Fill missing values with the median of the column
median_total_bedrooms = california_df["total_bedrooms"].median()
california_df["total_bedrooms"].fillna(median_total_bedrooms, inplace=True)

# Check if missing values were filled
print("Missing values filled with median:")
print(california_df["total_bedrooms"].isnull().sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

# Fetch the California housing dataset
california_housing = fetch_california_housing()

# Convert the dataset into a pandas DataFrame
california_df = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)

# Introduce artificial missing values in the "total_bedrooms" column
california_df.loc[2:5, "total_bedrooms"] = None

# Fill missing values using linear interpolation
california_df["total_bedrooms_interpolated"] = california_df["total_bedrooms"].interpolate(method="linear")

# Plot the original and interpolated values
plt.figure(figsize=(10, 6))

# Plot original values
plt.plot(california_df.index, california_df["total_bedrooms"], marker='o', linestyle='-', label='Original')

# Plot interpolated values
plt.plot(california_df.index, california_df["total_bedrooms_interpolated"], marker='o', linestyle='-', label='Interpolated')

plt.xlabel('Index')
plt.ylabel('Total Bedrooms')
plt.title('Interpolation of Missing Values in "total_bedrooms" Column')
plt.legend()
plt.grid(True)
plt.show()
```

# Choose a different dataset and handle null values using different techniques. Be creative!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Introduce artificial missing values
np.random.seed(0)  # for reproducibility
missing_mask = np.random.rand(*iris_df.shape) < 0.1  # 10% missing values
iris_df[missing_mask] = np.nan

# Visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(iris_df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title('Missing Values in Iris Dataset')
plt.show()

# Mean imputation
iris_df_mean_imputed = iris_df.fillna(iris_df.mean())

# Median imputation
iris_df_median_imputed = iris_df.fillna(iris_df.median())

# Interpolation
iris_df_interpolated = iris_df.interpolate(method='linear')

# Visualize the effect of each technique on a specific feature (e.g., sepal length)
feature_to_visualize = 'sepal length (cm)'
plt.figure(figsize=(12, 6))
plt.plot(iris_df.index, iris_df[feature_to_visualize], label='Original', marker='o')
plt.plot(iris_df_mean_imputed.index, iris_df_mean_imputed[feature_to_visualize], label='Mean Imputed', linestyle='--', marker='s')
plt.plot(iris_df_median_imputed.index, iris_df_median_imputed[feature_to_visualize], label='Median Imputed', linestyle='-.', marker='^')
plt.plot(iris_df_interpolated.index, iris_df_interpolated[feature_to_visualize], label='Interpolated', linestyle=':', marker='x')
plt.xlabel('Sample Index')
plt.ylabel('Feature Value')
plt.title(f'Effect of Imputation Techniques on "{feature_to_visualize}" Feature')
plt.legend()
plt.grid(True)
plt.show()
```
